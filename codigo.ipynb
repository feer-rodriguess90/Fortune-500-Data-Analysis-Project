```python
# Check the dataset for missing values
# Leverage the isnull() and sum() functions to find the number of missing values in each column
missing_values = df.isnull().sum()

# urn the missing value counts into percentages
missing_values_percentage = (missing_values / len(df)) * 100

# Return counts and percentages of missing values in each column
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage (%)': missing_values_percentage})
missing_data.sort_values(by='Percentage (%)', ascending=False)
```

Here's an overview of the missing values in the dataset: <br>
 **Newcomer to the Fortune 500:** Companies not included in the Fortune 500 list for the given year, incomplete data extraction or scraping process, or data entry errors. <br>
 **Profits Percent Change, Revenue Percent Change:** Lack of historical data for some companies, especially if they are newly listed or if the data is not available for the specified period. <br>
 **Change in Rank, Dropped in Rank, Gained in Rank:** Companies newly included in the Fortune 500 list for the current year, missing historical rank data, or data discrepancies. <br>
**Ticker, MarketCap March 31, MarketCap Updated** They may not be available for all companies, especially if they are privately held or if the data source does not provide this information, or it may be missing or incomplete in the dataset. <br>
> * Missing values in the dataset can arise due to various factors such as data availability, extraction methods, and inherent limitations in data sources.


```python
#Numeric Features 
#Replace the missing values with '0' using 'fiilna' method
df['Change_in_Rank'] = df['Change_in_Rank'].fillna(0)
df['MarketCap_March31_M'] = df['MarketCap_March31_M'].fillna(0)
df['ProfitsPercentChange'] = df['ProfitsPercentChange'].fillna(0)
df['MarketCap_Updated_M'] = df['MarketCap_Updated_M'].fillna(0)
df['RevenuePercentChange'] = df['RevenuePercentChange'].fillna(0)
#Categorical Features 
#Replace the missing values with 'NA' using 'fiilna' method
df['Ticker'] = df['Ticker'].fillna('NA')
df['Gained_in_Rank']=df['Gained_in_Rank'].fillna('NA')
df['Dropped_in_Rank']=df['Dropped_in_Rank'].fillna('NA')
df['Newcomer_to_the_Fortune500']=df['Newcomer_to_the_Fortune500'].fillna('NA')
df['Growth_in_Jobs']=df['Growth_in_Jobs'].fillna('NA')
df['CEO']=df['CEO'].fillna('NA')
df['Footnote']=df['Footnote'].fillna('NA')
df.isna().sum()
```

```python
# Check for Duplicates after replacement
print("Duplicate Rows:")
df.duplicated().sum()
```

```python
#Here we can see the most popular Sectors
sectors = df['Sector'].value_counts()

plt.figure(figsize=(12, 7))
sns.barplot(x=sectors.index, y=sectors.values, hue=sectors.index, palette="magma", legend=False)
plt.title('Most Popular Sectors', fontsize=16)
plt.xlabel('Sectors', fontsize=14)
plt.ylabel('Sectors Values', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The top 3 Sectors: 
> * Financials
> * Technology
> * Energy

```python
# Here we can see the percentage of profitable companies

# Calculate the count of profitable and non-profitable companies
profitable_counts = df['Profitable'].value_counts()

labels = profitable_counts.index
plt.figure(figsize=(8, 6))
plt.pie(profitable_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['rebeccapurple', 'burlywood'], explode=[0.05, 0.05])
plt.title('Percentage of Profitable Companies')
plt.legend()
plt.show()
```

> * The majority of companies in the dataset are profitable, with 85.9% reporting positive profits. This indicates a healthy business environment overall.
> * The 14.1% of companies that are not profitable may face challenges such as declining revenue, high expenses, or market disruptions. They may need to reassess their business strategies, operations, and cost structures to improve profitability and long-term sustainability.
> * Non-profitable companies can leverage insights from profitable peers to identify areas for improvement and implement strategies to enhance financial performance. This may include cost-cutting measures, product innovation, or market expansion efforts.

```python
# Here we can see the percentage of Female CEO

# Calculate the count of female and male CEO
profitable_counts = df['FemaleCEO'].value_counts()

labels = profitable_counts.index
plt.figure(figsize=(8, 6))
plt.pie(profitable_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['rebeccapurple', 'burlywood'], explode=[0.05, 0.05])
plt.title('Percentage of Female CEO')
plt.legend()
plt.show()
```

```python

# Group the data by sector and type of growth in jobs
sector_growth = df.groupby(['Sector', 'Growth_in_Jobs']).size().unstack()

# Plot the stacked bar chart
sector_growth.plot(kind='bar', stacked=True, figsize=(10, 6), color = ('purple', 'thistle', 'blueviolet'))
plt.title('Sector-wise Growth in Jobs')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.legend(title='Growth in Jobs', loc='upper right')
plt.show()
```

Understanding the factors contributing to job growth in these sectors can inform workforce development strategies, education initiatives, and talent acquisition efforts to support economic growth and innovation in these critical industries. 
> * The financials sector experiencing the highest growth in jobs indicates potential expansion and increased demand for financial services, such as banking, investment, and insurance.
> * The technology sector's significant growth in jobs reflects the industry's pivotal role in driving innovation, digital transformation, and economic development.
> * Despite facing challenges such as market volatility and environmental concerns, the energy sector's growth in jobs indicates ongoing demand for energy-related services and initiatives.
